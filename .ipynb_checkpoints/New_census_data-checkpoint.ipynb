```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pathlib import Path

#read in all files
jobs_2022 = Path('Resources_2/2022_jobsdata.csv')
jobs_2021 = Path('Resources_2/2021_jobsdata.csv')
jobs_df_1 = pd.DataFrame(pd.read_csv(jobs_2022))
jobs_df2 = pd.DataFrame(pd.read_csv(jobs_2021))


```

```python
#filter for total employee numbers for zip code 
jobs = jobs_df_1[['Label (Grouping)','Civilian employed population 16 years and over']]
jobs_df = pd.DataFrame(jobs)
jobs_df.head(33)

#create empty lists for getting the zip data and the total employment data
zips = []

totals = []

#this for loops looks for specific index to find the zip code and adds to the empty established list
for i in range(0, len(jobs_df), 16):
    new_zip = jobs_df.iloc[i,0]
    zips.append(new_zip)

#this for loop looks for specific index to find the total employment number change and adds tot he empty established list
for i in range(2, len(jobs_df), 16):
    new_total = jobs_df.iloc[i,1]
    totals.append(new_total)

zips
totals

#turn the zips and totals data into a full dataframe
full_df = pd.DataFrame({'Zip Code': zips,
                        'Total Employed 2022': totals})
full_df

#clean up zip code data
full_df['Zip Code'] = full_df['Zip Code'].str.replace('ZCTA5 ', '')
full_df

#clean up employed data and turn the data into integer for calculation
full_df['Total Employed 2022'] = full_df['Total Employed 2022'].str.replace(',','')
full_df['Total Employed 2022'] = full_df['Total Employed 2022'].astype(int)

full_df

```

```python
jobs2 = jobs_df2[['Label (Grouping)','Civilian employed population 16 years and over']]
jobs_df2 = pd.DataFrame(jobs2)
jobs_df2.head(33)

zips = []

totals = []

for i in range(0, len(jobs_df2), 16):
    new_zip = jobs_df2.iloc[i,0]
    zips.append(new_zip)

for i in range(2, len(jobs_df2), 16):
    new_total = jobs_df2.iloc[i,1]
    totals.append(new_total)

zips
totals

full_df2 = pd.DataFrame({'Zip Code': zips,
                        'Total Employed 2021': totals})
full_df2

full_df2['Zip Code'] = full_df2['Zip Code'].str.replace('ZCTA5 ', '')
full_df2

full_df2['Total Employed 2021'] = full_df2['Total Employed 2021'].str.replace(',','')
full_df2['Total Employed 2021'] = full_df2['Total Employed 2021'].astype(int)

full_df2
```

```python
comp_df = pd.merge(full_df, full_df2,on='Zip Code')
change = comp_df['Total Employed 2022'] / comp_df['Total Employed 2021'] - 1

comp_df['Employment Change 2022'] = (change*100).map('{:.2f}%'.format)
comp_df

all_zips = pd.read_csv('Resources_2/full_zip.csv')
all_zips_df = pd.DataFrame(all_zips)


comp_df.rename(columns={"Zip Code": "Zip"}, inplace=True)

comp_df['Zip'] = comp_df['Zip'].astype(int)

new_df = pd.merge(all_zips_df,comp_df,on='Zip',how='inner')

new_df = new_df.loc[new_df['Total Employed 2021'] != 0,:]
new_df = new_df.loc[new_df['Total Employed 2021'] > 350,:]

new_df.sort_values('Employment Change 2022', ascending=False)

new_df

```

```python
new_df['Employment Change 2022'] = new_df['Employment Change 2022'].str.replace('%','')
new_df['Employment Change 2022'] = new_df['Employment Change 2022'].astype(float)
new_df['2022'] = new_df['2022'].str.replace('%','')
new_df['2022'] = new_df['2022'].astype(float)
```

```python
new_df['Employment Change 2022'] = new_df['Employment Change 2022'] / 100
new_df['2022'] = new_df['2022'] / 100



scatter_df = new_df.sort_values('Employment Change 2022')
plt.scatter(scatter_df['2022'],scatter_df['Employment Change 2022'],marker='o',facecolors='blue',edgecolors='black')

# Incorporate the other graph properties
plt.title('Population vs Employment')
plt.xlabel('Population Change %')
plt.ylabel('Employment Change %')
plt.grid(color='black')

# Save the figure

# Show plot
plt.show()
```

```python
# Linear regression analysis
pop = scatter_df['2022']
employment = scatter_df['Employment Change 2022']
p_slope, p_int, p_r, p_p, p_std_err = linregress(pop, employment)
p_fit = p_slope * pop + p_int
line_eq = "y = " + str(round(p_slope,2)) + "x + " + str(round(p_int,2))

plt.scatter(pop,employment,marker='o',facecolors='blue',edgecolors='black')
plt.plot(pop,p_fit,"-",color='red')
plt.annotate(line_eq,(-0.3,0.4),fontsize=15,color='red')
plt.xlabel('Population Change %')
plt.ylabel('Employment Change %')
plt.grid(color='black')
plt.savefig('Images/Fig5')
plt.show()
print(f'The r value is {p_r}')
```

```python
#bar graph for top 25
new_df

top_employed_df = new_df.sort_values('Employment Change 2022',ascending=False)
top_employed_df = top_employed_df.head(25)

top_employed_df

plt.bar(top_employed_df['City'],top_employed_df['Employment Change 2022'],color='green',alpha=.6)
plt.title('Top 25 Zip Codes by Employment Change')
plt.xticks(rotation=85)
plt.ylabel('Employment Change')
plt.subplots_adjust(bottom=0.3)
plt.savefig('Images/Fig6')
plt.show()
```
