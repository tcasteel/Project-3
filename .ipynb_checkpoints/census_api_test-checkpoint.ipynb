```python
#geographical mobility in the past year in the united states

url = 'https://api.census.gov/data/2022/acs/acsse?get=NAME,K200701_001E&for=place:*&key=' + {api_key}

response = requests.get(url)

json_response = response.json()

geo_data = pd.DataFrame(json_response[1:], columns=json_response[0])

geo_data = geo_data.rename(columns={'K200701_001E': 'Geographical Mobility total'})

geo_data = geo_data.dropna()

geo_data

geo_data['Geographical Mobility total'] = geo_data['Geographical Mobility total'].astype(int)

geo_data = geo_data.sort_values('Geographical Mobility total', ascending=False)

geo_data.head(30)
# geo_data.loc[(geo_data['state'] == '35')]
```

```python
tx = geo_data.loc[(geo_data['state']) == '48']

tx.info()
```

```python
#pull API data by industry
agrurl = 'https://api.census.gov/data/2022/acs/acsse?get=NAME,K202403_002E&for=place:*&key=' + {api_key}

agresponse = requests.get(agrurl)

agrjson_response = agresponse.json()

agr_data = pd.DataFrame(agrjson_response[1:], columns=agrjson_response[0])

agr_data = agr_data.rename(columns={'K202403_002E': 'Total Employed'})

agr_data = agr_data.dropna()

agr_data = agr_data.loc[(agr_data['Total Employed']) != '0']

agr_data['Total Employed'] = agr_data['Total Employed'].astype(int)

agr_data22 = agr_data.sort_values('Total Employed', ascending=False)

agr_data22['Industry'] = 'Agriculture'
agr_data22['Year'] = '2022'
agr_data22 = agr_data22[['NAME', 'Industry', 'Year', 'Total Employed']]
agr_data22 = agr_data22.head(20)
agr_data22
```

```python
#pull API URL
paurl = 'https://api.census.gov/data/2022/acs/acsse?get=NAME,K202403_014E&for=place:*&key=' + {api_key}

#do request for pull
paresponse = requests.get(paurl)

#convert request to json
pajson_response = paresponse.json()

#turn json response into a dataframe and adjust the second row to the column headers
pa_data = pd.DataFrame(pajson_response[1:], columns=pajson_response[0])

#rename column for group name we are pulling data frame to a clear column name
pa_data = pa_data.rename(columns={'K202403_014E': 'Total Employed'})

#drop na
pa_data = pa_data.dropna()

#filter out cities with 0 
pa_data = pa_data.loc[(pa_data['Total Employed']) != '0']

#Change estimate totals to integer
pa_data['Total Employed'] = pa_data['Total Employed'].astype(int)

#Sort the Estimate totals
pa_data22 = pa_data.sort_values('Total Employed', ascending=False)

pa_data22['Industry'] = 'Public Administration'
pa_data22['Year'] = '2022'
pa_data22 = pa_data22[['NAME', 'Industry', 'Year', 'Total Employed']]
pa_data22 = pa_data22.head(20)
pa_data22
```
