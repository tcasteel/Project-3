```python
import pandas as pd
from pathlib import Path
```

```python
cities10s = Path('Resources_3/cities_no_zips_10s.csv')
cities20s = Path('Resources_3/cities_no_zips_20s.csv')
```

```python
data_2010s = pd.DataFrame(pd.read_csv(cities10s))
data_2020s = pd.DataFrame(pd.read_csv(cities20s))
```

```python
data_2010s = data_2010s.iloc[3:]
data_2010s = data_2010s.iloc[:-5]
from_17_to_19 = ['table with row headers in column A and column headers in rows 3 through 4 (leading dots indicate sub-parts)', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12']
data_2010s = data_2010s[from_17_to_19]
```

```python
data_2020s = data_2020s.iloc[3:]
data_2020s = data_2020s.iloc[:-5]
from_20_to_22 = ['table with row headers in column A and column headers in rows 3 through 4 (leading dots indicate sub-parts)', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4']
data_2020s = data_2020s[from_20_to_22]
```

```python
data_2010s = data_2010s.rename(columns = {'table with row headers in column A and column headers in rows 3 through 4 (leading dots indicate sub-parts)': 'City, State', 'Unnamed: 10': '2017', 'Unnamed: 11': '2018', 'Unnamed: 12': '2019'})
data_2010s = data_2010s.drop_duplicates(subset='City, State', keep='first')
data_2010s = data_2010s.set_index('City, State')
data_2010s
```

```python
data_2020s = data_2020s.rename(columns = {'table with row headers in column A and column headers in rows 3 through 4 (leading dots indicate sub-parts)': 'City, State', 'Unnamed: 2': '2020', 'Unnamed: 3': '2021', 'Unnamed: 4': '2022'})
data_2020s = data_2020s.drop_duplicates(subset='City, State', keep='first')
data_2020s = data_2020s.set_index('City, State')
data_2020s
```

```python
city_population = pd.merge(data_2010s, data_2020s, on="City, State")
year_columns = ['2017', '2018', '2019', '2020', '2021', '2022']
city_population[year_columns] = city_population[year_columns].replace({',': ''}, regex=True).astype('int')
city_population = city_population[city_population['2017'] != 0]
city_population.to_csv('Resources_3/population_by_city.csv', index = True)
```

```python
city_population
```

```python
city_growth = pd.DataFrame(index = city_population.index)
city_growth['2017'] = city_population['2017']
city_growth['2018'] = city_population['2018']
city_growth['2019'] = city_population['2019']
city_growth['2020'] = city_population['2020']
city_growth['2021'] = city_population['2021']
city_growth['2022'] = city_population['2022']
original_year = 0
for year in range(original_year, len(city_population.columns) - 1):
    current_year = year + 1
    current_column = city_population.columns[current_year]
    percent_change = ((city_population[current_column] - city_population.iloc[:, year]) / city_population.iloc[:, year]) * 100
    city_growth[current_column] = percent_change.map("{:.2f}".format)
    
pd.reset_option('display.max_rows', None)
city_growth.replace("nan%", "0.00%", inplace = True)
city_growth = city_growth.rename(columns={'2017': '2017 Population', '2018': '2018 PctGrowth', '2019': '2019 PctGrowth', '2020': '2020 PctGrowth', '2021': '2021 PctGrowth', '2022': '2022 PctGrowth'})
city_growth = city_growth[city_growth['2017 Population'] != 0]
city_growth.to_csv('Resources_3/growth_by_city.csv', index = True)
```
