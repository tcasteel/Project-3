```python
import pandas as pd
from pathlib import Path
```

```python
population_2017 = Path("Resources_3/zip_pop_2017.csv")
population_2018 = Path("Resources_3/zip_pop_2018.csv")
population_2019 = Path("Resources_3/zip_pop_2019.csv")
population_2020 = Path("Resources_3/zip_pop_2020.csv")
population_2021 = Path("Resources_3/zip_pop_2021.csv")
population_2022 = Path("Resources_3/zip_pop_2022.csv")
cities = Path("Resources_3/zips_with_cities.csv")
```

```python
data_2017 = pd.DataFrame(pd.read_csv(population_2017))
data_2018 = pd.DataFrame(pd.read_csv(population_2018))
data_2019 = pd.DataFrame(pd.read_csv(population_2019))
data_2020 = pd.DataFrame(pd.read_csv(population_2020))
data_2021 = pd.DataFrame(pd.read_csv(population_2021))
data_2022 = pd.DataFrame(pd.read_csv(population_2022))
```

```python
city_zip = pd.DataFrame(pd.read_csv(cities, delimiter =";"))
city_zip = city_zip['Zip Code'].str.split(';', expand=True)
city_zip = city_zip.iloc[:, :4]
city_zip.columns = ['Zip Code', 'City', 'State', 'State Name']
city_zip['Zip Code'] = city_zip['Zip Code'].apply(lambda x: x.zfill(5))
city_zip = city_zip.set_index('Zip Code')
city_zip.to_csv("city_zip.csv", index = True)
```

```python
data_2017 = data_2017.drop(0).drop(columns=['GEO_ID', 'B01003_001M']).rename(columns={'NAME': 'Zip Code', 'B01003_001E': '2017'})
data_2018 = data_2018.drop(0).drop(columns=['GEO_ID', 'B01003_001M']).rename(columns={'NAME': 'Zip Code', 'B01003_001E': '2018'})
data_2019 = data_2019.drop(0).drop(columns=['GEO_ID', 'B01003_001M']).rename(columns={'NAME': 'Zip Code', 'B01003_001E': '2019'})
data_2020 = data_2020.drop(0).drop(columns=['GEO_ID', 'B01003_001M']).rename(columns={'NAME': 'Zip Code', 'B01003_001E': '2020'})
data_2021 = data_2021.drop(0).drop(columns=['GEO_ID', 'B01003_001M']).rename(columns={'NAME': 'Zip Code', 'B01003_001E': '2021'})
data_2022 = data_2022.drop(0).drop(columns=['GEO_ID', 'B01003_001M']).rename(columns={'NAME': 'Zip Code', 'B01003_001E': '2022'})
```

```python
data_2017['Zip Code'] = data_2017['Zip Code'].str.replace('ZCTA5 ', '')
data_2018['Zip Code'] = data_2018['Zip Code'].str.replace('ZCTA5 ', '')
data_2019['Zip Code'] = data_2019['Zip Code'].str.replace('ZCTA5 ', '')
data_2020['Zip Code'] = data_2020['Zip Code'].str.replace('ZCTA5 ', '')
data_2021['Zip Code'] = data_2021['Zip Code'].str.replace('ZCTA5 ', '')
data_2022['Zip Code'] = data_2022['Zip Code'].str.replace('ZCTA5 ', '')
```

```python
data_2017 = data_2017.set_index('Zip Code')
data_2018 = data_2018.set_index('Zip Code')
data_2019 = data_2019.set_index('Zip Code')
data_2020 = data_2020.set_index('Zip Code')
data_2021 = data_2021.set_index('Zip Code')
data_2022 = data_2022.set_index('Zip Code')
```

```python
first_merge = pd.merge(city_zip, data_2017, on = 'Zip Code')
second_merge = pd.merge(first_merge, data_2018, on = 'Zip Code')
third_merge = pd.merge(second_merge, data_2019, on = 'Zip Code')
fourth_merge = pd.merge(third_merge, data_2020, on = 'Zip Code')
fifth_merge = pd.merge(fourth_merge, data_2021, on = 'Zip Code')
us_zip_pop = pd.merge(fifth_merge, data_2022, on = 'Zip Code')
us_zip_pop = us_zip_pop.sort_index()
us_zip_pop.to_csv('us_zip_pop.csv', index = True)
us_zip_pop
```

```python
us_city_pop = us_zip_pop.reset_index()
us_city_pop = us_city_pop.drop(columns = 'Zip Code')
us_city_pop['City, State'] = us_city_pop['City'] + ', ' + us_city_pop['State Name']
us_city_pop = us_city_pop[['City, State', 'State', '2017', '2018', '2019', '2020', '2021', '2022']]
us_city_pop[['2017', '2018', '2019', '2020', '2021', '2022']] = us_city_pop[['2017', '2018', '2019', '2020', '2021', '2022']].apply(pd.to_numeric, errors='coerce')
us_city_pop = us_city_pop.sort_values("City, State")
```

```python
us_city_total = us_city_pop.groupby('City, State').sum().reset_index()
us_city_total['State'] = us_city_total['State'].str[:2]
us_city_total.to_csv('us_city_total4.csv', index = False)
```
